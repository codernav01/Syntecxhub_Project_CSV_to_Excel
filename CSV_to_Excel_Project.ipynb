```python
import pandas as pd

# Full path use karo (kyunki file D drive me hai)
file_path = "D:/Syntecxhub_Project_CSV_to_Excel/sample.csv"

# CSV read karo (tumhara file tab separated hai)
df = pd.read_csv(file_path, sep="\t")

print("Original Data:")
print(df)
```

```python
# Fill missing age with 0
df["age"] = df["age"].fillna(0)

# Convert date safely (invalid date -> NaT)
df["date"] = pd.to_datetime(df["date"], errors="coerce")

# Rename column
df.rename(columns={"name": "Full_Name"}, inplace=True)

print("Cleaned Data:")
print(df)
```

```python
output_path = "D:/Syntecxhub_Project_CSV_to_Excel/output.xlsx"

df.to_excel(output_path, index=False)

print("ðŸŽ‰ Project Completed Successfully!")
print("File saved at:", output_path)
```
